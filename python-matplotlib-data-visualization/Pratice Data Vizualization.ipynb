```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
plt.plot(years,apples)
plt.plot(years,oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.legend(['Apples','Oranges'])
plt.title('Crop yeild in kanto');
```

```python
plt.plot(years,apples,marker='o',c='red',ls=':',mfc='cyan')
plt.plot(years,oranges,marker='<',c='k',lw=1,ms=8,mew=2,mfc='red',mec='yellow')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title('Crops yield in kanto')
plt.legend(['Apples','Oranges']);
```

```python
# using 'fmt argument' ---> which allows you to style the line using shorthand.
plt.figure(figsize=(10,6))
sns.set_style("whitegrid")
plt.plot(years,apples,'o:r')
plt.plot(years,oranges,'>-.k')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title('Crops yield in kanto')
plt.legend(['Apples','Oranges']);
```

```python

flights_df = sns.load_dataset("flights")
```

```python
flights_df
```

```python
# how to put the column "where i want to " in the dataframe.
# This is the way when you have a lots of columns.

col_to_move = flights_df['month']

flights_df = flights_df.drop('month',axis=1)

flights_df.insert(0,'month',col_to_move)

# Other way
# When you have less columns
flights_df = flights_df[['month',"year","passengers"]]
```

```python
flights_df
```

```python
flights_df = flights_df.pivot(columns=['year'],index=['month'],values=['passengers'])
```
